# Exploratory Data Analysis on the English Premier League

```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
#loading the dataset
df = pd.read_csv("https://raw.githubusercontent.com/BernardOforiBoateng/Data_Wrangling_and_EDA_DSCI401/main/Project/premier-league-matches.csv")
df.head()
```

```python
#Rename Columns
df.rename(columns={"Wk": "Week", "Home": "HomeTeam", "Away": "AwayTeam", "FTR" :"FullTimeResult"}, inplace=True)
df.head()
```

```python
#Create a points column
# A win is 3 points, draw is 1 point each
df['Points']=3
df.loc[df['FullTimeResult']=='D', 'Points']=1
df.head()
```

```python
#Lets check the total number of games played in each season
df['Season_End_Year'].value_counts()
```

As we can see the premier league originally started with 22 teams for the first three seasons after which it was changed to 20 teams.
The Premier League reduced the number of teams from a 22 team league with 462 matches a season to  a 20 team league to combat fixture congestion and player fatigue, to improve the quality of matches and to increase its competitiveness both domestically and internationally.

```python
#Get unique team names from the dataset
home_teams = df['HomeTeam'].unique()
away_teams = df['AwayTeam'].unique()
all_teams = pd.Series(list(set(home_teams) | set(away_teams)))

# Display unique team names
all_teams
```

Now the Premier League much like most European Football leagues are dominated by a select few who have the financial power to buy the best players and employ the best coaches. Lets check the trend of how the "Big Six" teams have perfomed over the years.

```python
#Rename specific team name in the dataset to full name
df['HomeTeam'] = df['HomeTeam'].replace("Nott'ham Forest", "Nottingham Forest")
df['AwayTeam'] = df['AwayTeam'].replace("Nott'ham Forest", "Nottingham Forest")

#Get unique team names from the dataset
home_teams = df['HomeTeam'].unique()
away_teams = df['AwayTeam'].unique()
all_teams = pd.Series(list(set(home_teams) | set(away_teams)))

#Create a new DataFrame to store total points for each team at the end of each season
total_points_df = pd.DataFrame(index=all_teams, columns=range(1993, 2024))
total_points_df = total_points_df.fillna(0)

#Iterate through each row in the original DataFrame
for index, row in df.iterrows():
    season_end_year = row['Season_End_Year']
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    full_time_result = row['FullTimeResult']

    # Update points based on match outcome
    if full_time_result == 'H':  #Home team won
        total_points_df.at[home_team, season_end_year] += 3
    elif full_time_result == 'A':  #Away team won
        total_points_df.at[away_team, season_end_year] += 3
    else:  #Draw
        total_points_df.at[home_team, season_end_year] += 1
        total_points_df.at[away_team, season_end_year] += 1

#Display the total points for each team at the end of each season
total_points_df
```

```python
#List of the big six teams
big_six_teams = ['Manchester Utd', 'Liverpool', 'Arsenal', 'Chelsea', 'Manchester City', 'Tottenham']

#Get  data for the big six teams
big_six_data = total_points_df.loc[big_six_teams]

#Transpose the DataFrame for better visualization
big_six_data = big_six_data.transpose()

#Create an interactive line plot using Plotly Express
fig = px.line(big_six_data, x=big_six_data.index, y=big_six_teams,
              labels={'index': 'Season', 'value': 'Total Points'},
              title='Performance of Big Six Teams Over the Years',
              line_shape="linear",
              )

#Add gridlines and show the plot
fig.update_layout(showlegend=True, xaxis=dict(showgrid=True), yaxis=dict(showgrid=True))
fig.show()
```

Teams with the Most Home and Away Points

```python
#Calculate home and away points for each team in each season
home_points = df[df['FullTimeResult'].isin(['H', 'D'])].groupby(['Season_End_Year', 'HomeTeam'])['Points'].sum().reset_index()
away_points = df[df['FullTimeResult'].isin(['A', 'D'])].groupby(['Season_End_Year', 'AwayTeam'])['Points'].sum().reset_index()

#Team with the most home points for each season
most_home_points = home_points.groupby('Season_End_Year').apply(lambda x: x.loc[x['Points'].idxmax()]).reset_index(drop=True)

#Team with the most away points for each season
most_away_points = away_points.groupby('Season_End_Year').apply(lambda x: x.loc[x['Points'].idxmax()]).reset_index(drop=True)
```

```python
#Display the results
print("Team with the Most Home Points Each Season:")
most_home_points
```

```python
print("Team with the Most Away Points Each Season:")
most_away_points
```

Teams with the Least amount of points in each season

```python
#Calculate home and away points for each team in each season
home_points = df[df['FullTimeResult'].isin(['H', 'D'])].groupby(['Season_End_Year', 'HomeTeam'])['Points'].sum().reset_index()
away_points = df[df['FullTimeResult'].isin(['A', 'D'])].groupby(['Season_End_Year', 'AwayTeam'])['Points'].sum().reset_index()

#Merge home and away points, filling NaN with 0
total_points = pd.merge(home_points, away_points, how='outer', left_on=['Season_End_Year', 'HomeTeam'], right_on=['Season_End_Year', 'AwayTeam'])
total_points = total_points.fillna(0)

#Calculate total points for each team in each season
total_points['TotalPoints'] = total_points['Points_x'] + total_points['Points_y']

#Get the team with the lowest total points in each season
lowest_points_per_season = total_points.groupby('Season_End_Year').apply(lambda x: x.loc[x['TotalPoints'].idxmin()]).reset_index(drop=True)

#Display the results
print("Teams with the Lowest Total Points Accumulated in Each Season (Including Last Season):")
print(lowest_points_per_season[['Season_End_Year', 'HomeTeam', 'Points_x', 'Points_y', 'TotalPoints']])
```

As seen, Derby County holds the undesirable record for the lowest points ever accumulated in a Premier League season, accruing only 11 points. This achievement, or rather lack thereof, occurred during the 2007-2008 season.

```python
# Get the team with the highest total points in each season
title_winners = total_points.groupby('Season_End_Year')['HomeTeam'].apply(lambda x: x[total_points.loc[x.index, 'TotalPoints'].idxmax()]).reset_index()
title_winners.columns = ['Season_End_Year', 'Team']

# Count the number of titles for each team
title_counts = title_winners['Team'].value_counts().reset_index()
title_counts.columns = ['Team', 'count']

# Display the results
print("Teams with Title Counts:")
title_counts
```

```python
# Create a pie chart
fig = px.pie(title_counts, names='Team', values='count', title='Premier League Title Counts by Team')
fig.show()
```

Goals

Analyze the distribution of goals scored in each season

```python
#Convert 'Date' to datetime data type
df['Date'] = pd.to_datetime(df['Date'])

#Calculate total goals scored in each season
total_goals_by_season_home = df.groupby('Season_End_Year')['HomeGoals'].sum().reset_index()
total_goals_by_season_away = df.groupby('Season_End_Year')['AwayGoals'].sum().reset_index()

#Calculate total goals (home + away)
total_goals_by_season = total_goals_by_season_home.merge(total_goals_by_season_away, on='Season_End_Year', suffixes=('_home', '_away'))
total_goals_by_season['TotalGoals'] = total_goals_by_season['HomeGoals'] + total_goals_by_season['AwayGoals']

#Create plot
fig = px.bar(total_goals_by_season, x='Season_End_Year', y='TotalGoals', text='TotalGoals',
             labels={'TotalGoals': 'Total Goals'}, title='Total Goals Scored in Each Season')

#Set labels and title
fig.update_layout(xaxis_title='Season End Year', yaxis_title='Total Goals', uniformtext_minsize=8, uniformtext_mode='hide')

#Show plot
fig.show()
```

```python
#Calculate the total goals across all seasons
total_goals_all_seasons = total_goals_by_season['TotalGoals'].sum()

#Display the total amount of goals
print("Total Goals Scored Across All Seasons:", total_goals_all_seasons)
```

```python
# Calculate home and away points for each team in each season
home_points = df[df['FullTimeResult'].isin(['H', 'D'])].groupby(['Season_End_Year', 'HomeTeam'])['Points'].sum().reset_index()
away_points = df[df['FullTimeResult'].isin(['A', 'D'])].groupby(['Season_End_Year', 'AwayTeam'])['Points'].sum().reset_index()

# Merge home and away points, filling NaN with 0
total_points = pd.merge(home_points, away_points, how='outer', left_on=['Season_End_Year', 'HomeTeam'], right_on=['Season_End_Year', 'AwayTeam'])
total_points = total_points.fillna(0)

# Calculate total points for each team in each season
total_points['TotalPoints'] = total_points['Points_x'] + total_points['Points_y']

# Get the 3 teams with the lowest total points in each season
lowest_points_per_season = (
    total_points.sort_values(['Season_End_Year', 'TotalPoints'])
    .groupby('Season_End_Year').head(3)
    .reset_index(drop=True)
)
```

```python
#Count the occurrences of each team in the list of teams with the lowest points
teams_count = lowest_points_per_season['HomeTeam'].value_counts().reset_index()
teams_count.columns = ['Team', 'Count']

#Get the top 5 teams that appear the most
top_5_teams = teams_count.head(5)

print("Top 5 Teams With the most Relegations:")
top_5_teams
```
